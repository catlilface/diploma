```python
import pandas as pd
from tqdm import tqdm
parquet_filenames = [f'cv.{i}.parquet' for i in range(1, 12)]

clean_cv_df = pd.DataFrame()
for file in tqdm(parquet_filenames):
  chunk = pd.read_parquet(f'data/processed/{file}')
  moscow_chunk = chunk.query('stateRegionCode == "7700000000000"')
  clean_cv_df = pd.concat([clean_cv_df, moscow_chunk])

clean_cv_df.describe()
```

```python
clean_cv_df['educationList'].describe()
```

```python
clean_cv_df.to_parquet('data/clean/cv.parquet', compression='zstd', engine='pyarrow')
```
