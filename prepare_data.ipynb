# Гистограммы и диаграммы рассеяния

```python
def find_name(code, json_data):
    for item in json_data:
        try:
            if item['code'] == code:
                    if (item['type'] in ['int', 'bool', 'boolean']):
                        return item['name'], item['type']
        except:
            return False
    return False
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import json
from tqdm import tqdm

fig, axs = plt.subplots(4, 5, figsize=(26,16))
plt.subplots_adjust(wspace=0.4, hspace=0.4)

with open('./data/vacancies_attributes.json', 'r') as f:
    attr_data = json.load(f)

    i = 0
    for attr in tqdm(list(vacancies_df.columns)):
        if (not find_name(attr, attr_data)):
            continue

        attr_name, attr_type = find_name(attr, attr_data)
        sns.histplot(vacancies_df[attr], ax=axs[i // 5][i % 5])
        axs[i // 5][i % 5].set_title(attr_name)
        i += 1

print('Rendering', i)
```

## Лемматизация датасета с вакансиями

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("russian")

def lemmatize_text(text: str):
    if pd.isna(text):
        return None
    
    text = text.lower().split('-')

    return ' '.join(stemmer.stem(word) for word in text if word != '')
```

```python
lemma_vacancies_df = clean_vacancies_df.copy()

for column in tqdm(lemma_vacancies_df.columns):
    if 'string' in str(lemma_vacancies_df[column].dtype):
        lemma_vacancies_df[column] = lemma_vacancies_df[column].apply(lambda x: lemmatize_text(x))

lemma_vacancies_df.head()
```
